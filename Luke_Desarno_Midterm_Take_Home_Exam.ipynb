```python
D = np.array([16,12,29,5,9])    # These numbers are the diameter
C = np.array([44,31,53,19,14])  # These numbers are the circumference, They are for all the objects listed
```

```python
plt.plot(D,C,'g.') #plt.plot creates plot, you specify color with letter -> g = green, when followed with a period it will ne marked with dots on the chart
```

```python
x = D # D = diameter|
y = C # C = circumference
```

```python
m_guess = 0 #set equal to zero makes them have no value yet
b_guess = 0 
```

```python
x
```

```python
m_guess * x
```

```python
m_guess * x + b_guess
```

```python
y_guess = m_guess * x + b_guess #This is y=mx+b
```

```python
y_guess
```

```python
best_loss = 1000000000 #number of answers provided
best_m = 0
best_b = 0 

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        print("Now Testing: ", m_guess, b_guess) #Tested answers or "guesses" will be provided below
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2) #negative values will become positive

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r'); #r plots red lines
plt.plot(D,C,'g.'); #g with period after plots green dots
plt.show()

print(best_loss,best_m,best_b)
```
